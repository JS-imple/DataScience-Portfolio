```python
def get_match_data(region, match_id, api_key):
    api_url = (
        'https://' +
        region +
        '.api.riotgames.com/lol/match/v5/matches/' +
        match_id +
        '?api_key=' +
        api_key
    )

    resp = requests.get(api_url)
    data = resp.json()
    return data


        
```

```python
## extract win data from puuid data ##
```

```python
## function to check in player data if win is true/false
def did_win(puuid, match_data):
    part_index = match_data['metadata']['participants'].index(puuid)
    return match_data['info']['participants'][part_index]['win']
```

```python
## function to build URL, grab mataches of player
def get_matches(region, puuid, count, api_key):
    api_url = (
        'https://' +
        region +
        '.api.riotgames.com/lol/match/v5/matches/by-puuid/' +
        puuid +
        '/ids' +
        '?type=ranked' +
        '&start=0&' +
        'count=' +
        str(count) +
        '&api_key=' +
        api_key
    )
    while True:
        resp = requests.get(api_url)
    
        if resp.status_code == 429:
            print('Sleeping')
            time.sleep(10)
            continue
        
        
        data = resp.json()
        return data
```
